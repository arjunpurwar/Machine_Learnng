```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples= 200, n_features= 2, centers=4, random_state=101) # creating your own dataset
```

```python
plt.scatter(data[0][:,0], data[0][:,1])
```

```python
data[1] # y parameter - donot it in real world 
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_)
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1])
```

# How to find the optimal number of clusters

```python
kmeans.inertia_
```

```python
from numpy import random
import numpy as np
Ks = 11

mse = np.zeros((Ks-1))

for n in range(1, Ks):
  kclusters = n

  kmeans = KMeans(n_clusters = kclusters, random_state =0, n_init =15).fit(data[0])

  mse[n-1] = kmeans.inertia_  

plt.plot(range(1,Ks), mse)
plt.xlabel('Number of clusters')

plt.ylabel('MSE')

plt.title('Optimal K selection')
```
